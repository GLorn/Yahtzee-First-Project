# Yahtzee Functions

## First Dice Roll

```python
import random

def five_die():
    """Call for first 5 dice roll"""
    dice_roll = ["Your rolls:"]
    while len(dice_roll) < 6:
        dice_roll.append(random.randint(1,6))
    return dice_roll
```

## Ask Re-Roll

```python
def player_choose():
    """Gives the player a choice if they want to re roll their dice.
    out puts a bool"""
    print(dice_roll)
    roll_choice = input("\nDo you want to re-roll? Y/N  ").upper()
    if roll_choice == "Y":
        return True
    else:
        return False
```

## Updating Roll

```python
def update_roll(dice_roll, player_choice):
    """If the player chooses to re-roll.
    This function allows the player to select which dice to re-roll
    and updates the players dice roll."""
    if player_choice == True:
        dice_choice = int(input("Select which dice you want to re-roll (1-5)! Eg. 134. "))
        lst_choice = [int(i) for i in str(dice_choice)]
        print(f" Your choice is {lst_choice}")
        for value in lst_choice:
            dice_roll[value] = random.randint(1,6)
        return dice_roll
    else:
        return dice_roll
```

## Print Score Table

```python
from tabulate import tabulate

def table(game_score):
    """Called when we want to print the current score table to the players"""
    print(tabulate(game_score, headers=["Dice", "Player 1", "Player 2"]))
```

## Empty Score Table

```python
def initial_tables():
    """This is used to reset the score board for a new game."""
    lst_1 = ["1", "", ""]
    lst_2 = ["2", "", ""]
    lst_3 = ["3", "", ""]
    lst_4 = ["4", "", ""]
    lst_5 = ["5", "", ""]
    lst_6 = ["6", "", ""]
    list_of_lists = [lst_1, lst_2, lst_3, lst_4, lst_5, lst_6]
    
    space_table2 = ["", "", ""]
    title_table2 = ["Combinations", "Player 1", "Player 2"]
    dash_table2 = ["------------", "----------", "----------"]
    lst_10 = ["Chance", "", ""]
    lst_20 = ["Full House", "", ""]
    lst_30 = ["3 of a kind", "", ""]
    lst_40 = ["4 of a kind", "", ""]
    lst_50 = ["YAHTZEE", "", ""]
    list_of_lists_2 = [space_table2, title_table2, dash_table2, lst_10, lst_20, lst_30, lst_40, lst_50]
    
    global_list = list_of_lists + list_of_lists_2
    
    return global_list
```

## Dice Combos

### Chance

```python
dice_roll = ['Your rolls:', 1, 6, 3, 3, 6]
"""Function to check out-put the score if the player chooses to take a chance"""
def chance(dice_roll):
    chance_sum = sum(dice_roll[1:6])
    return chance_sum
```

### 3 of a Kind

```python
dice_roll = ['Your rolls:', 1, 3, 3, 3, 6]
def kind_3(dice_roll):
    """Function checks if the players dice roll is a 3 of a kind and out puts the score if True"""
    values = dice_roll[1:6]
    #values = values.sort()
    dct_count = {}
    
    for y in values:
        dct_count[y] = values.count(y)
                
    lst_dct_values = dct_count.values()
    seq_3 = [i for i in lst_dct_values if i >= 3]
    
    if seq_3:
        kind_3_sum = sum(dice_roll[1:6])
    else:
        kind_3_sum = 0
    return kind_3_sum
```

### 4 of a Kind

```python
dice_roll = ['Your rolls:', 1, 3, 3, 3, 3]
def kind_4(dice_roll):
    """Function checks if the players dice roll is a 4 of a kind and out puts the score if True"""
    values = dice_roll[1:6]
    #values = values.sort()
    dct_count = {}
    
    for y in values:
        dct_count[y] = values.count(y)
                
    lst_dct_values = dct_count.values()
    seq_4 = [i for i in lst_dct_values if i >= 4]
    
    if seq_4:
        kind_4_sum = sum(dice_roll[1:6])
    else:
        kind_4_sum = 0
    
    return kind_4_sum
```

### Full House

```python
dice_roll = ['Your rolls:', 1, 3, 3, 3, 1]
def full_house(dice_roll):
    """Function checks if the players dice roll is a Full House and out puts the score if True"""
    values = dice_roll[1:6]
    #values = values.sort()
    dct_count = {}
    
    for y in values:
        dct_count[y] = values.count(y)
                
    lst_dct_values = dct_count.values()
    seq_3 = [True for i in lst_dct_values if i == 3]
    seq_2 = [True for i in lst_dct_values if i == 2]
    
    if seq_3 and seq_2:
        return 25
    else:    
        return 0
```

### Yahtzee!

```python
dice_try = ['Your rolls:', 1, 2, 1, 1, 1]
def combo_yahtzee(dice_try):
    """Function checks if the players dice roll is Yahtzee (5 of a kind) and out puts the score if True"""
    values = dice_try[1:6]
    set_values = set(values)
    lst_set = list(set_values)
    if lst_set == [1] or lst_set == [2] or lst_set == [3] or lst_set == [4] or lst_set == [5] or lst_set == [6]:
        return 50
    else:
        return 0
```

## Which Table

```python
def which_table(): # upper or lower table to update, user should choose this
    """Allows the player to choose which table section they wish to input their score.
    1 for the upper table (Adding sums of single intergers)
    2 for the lower table (Special Combos)
    Returns selected value (1 or 2)"""
    selected_table = 0
    while selected_table not in range(1,3):
        try:
            selected_table = int(input("Select which table to add your score: 1 for Upper table, 2 for Lower table. \n"))
        except ValueError:
            selected_table = int(input("Select 1 or 2 only \n"))
    return selected_table
```

## Player Slot Choice

```python
def slot_choice(selected_table):
    """Allows the player to select which score slot to use for their turn
    function checks it is a valid value and outputs users choice
    (Takes the table choice as an imput)"""
    if selected_table == 1:
        user_table_choice = 0
        while user_table_choice not in range(1,7):
            try:
                user_table_choice = int(input("Select an empty score slot to update! (1-6): \n"))
            except ValueError:
                user_table_choice = int(input("Select a valid slot value between (1-6): \n"))
        return user_table_choice
    else:
        user_lower_table_choice = 0
        while user_lower_table_choice not in range(1,6):
            try:
                user_lower_table_choice = int(input("Select an empty score slot to update! (1-5): \n"))
                lower_index = user_lower_table_choice + 8
            except ValueError:
                user_lower_table_choice = int(input("Select a valid slot value between (1-5): \n"))
                lower_index = user_lower_table_choice + 8
        return lower_index
```

## Checking Table Slot

```python
def check_slot(user_table_choice, game_score, player):
    """Function checks the player selected slot to make sure it is empty to allow a score to be inputted
    if not it will ask the player again until a valid slot is chosen"""
    if player == "Player 1":
        player_index = 1
    else:
        player_index = 2
    if user_table_choice > 8:
        if game_score[user_table_choice][0] == "3 of a kind" or game_score[user_table_choice][0] == "Chance" or game_score[user_table_choice][0] == "Full House" or game_score[user_table_choice][0] == "4 of a kind" or game_score[user_table_choice][0] == "YAHTZEE":
            #print(f"i is: {user_table_choice}, and game_score: {game_score[user_table_choice][0]}")
            if game_score[user_table_choice][player_index] or game_score[user_table_choice][player_index] == 0:
                #print("Slot is taken")
                return False
            else:
                    #print("Slot is empty")
                    return True
    else:
        for i in range(6):
            if game_score[i][0] == str(user_table_choice):
                #print(f"i is: {i}, and game_score: {game_score[i][0]}")
                if game_score[i][player_index] or game_score[i][player_index] == 0:
                    #print("Slot is taken")
                    return False
                else:
                    #print("Slot is empty")
                    return True
```

## Checking End Game

```python
def game_end(game_score):
    """Function checks the board for empty slots (Game can only end if every slot is filled)
    Returns a bool"""
    for i in range(1,14):
        if check_slot(i, game_score, player):
               return False
    else:
        return True
```

### End Game Total Scores

```python
def total_scores(game_score):
    """Function adds up all the scores and announces the winner and both players scores"""
    
    player1_score = 0
    player2_score = 0
    
    for x in range(len(game_score)):
        #print(type(x))
        player1_score = game_score[0][1] + game_score[1][1] + game_score[2][1] + game_score[3][1] + game_score[4][1] + game_score[5][1] + game_score[9][1] + game_score[10][1] + game_score[11][1] + game_score[12][1] + game_score[13][1]
        player2_score = game_score[0][2] + game_score[1][2] + game_score[2][2] + game_score[3][2] + game_score[4][2] + game_score[5][2] + game_score[9][2] + game_score[10][2] + game_score[11][2] + game_score[12][2] + game_score[13][2]
        #player1_score += game_score[x][1]
        #player2_score += game_score[x][2]
     
    
    
    if player1_score > player2_score:
        print(f"\nPlayer 1 wins with a score of {player1_score}\nPlayer 2 scored {player2_score}")
    elif player1_score == player2_score:
        print(f"\nIts a tie! Both players scored {player1_score}")
    else:
        print(f"\nPlayer 2 wins with a score of {player2_score}\nPlayer 1 scored {player1_score}")
    
```

## Inputing Score to Table

```python
def player_column(player):
    """Function checks which player is playing to input the score into the correct column"""
    if player == "Player 1":
        player_col = 1
    else:
        player_col = 2
    return player_col
```

```python
def inputs_table(user_table_choice, dice_roll, player_col, game_score):
    """Function updates the score table with the players selected score.
    Takes as arguments - the player input, players column, current dice roll, and current score board
    """
    if user_table_choice <= 6:
        count_score = dice_roll.count(user_table_choice) * user_table_choice
        game_score[user_table_choice - 1][player_col] = count_score
    elif user_table_choice == 9:
        game_score[9][player_col] = chance(dice_roll)
    elif user_table_choice == 10:
        game_score[10][player_col] = full_house(dice_roll)
    elif user_table_choice == 11:
        game_score[11][player_col] = kind_3(dice_roll)
    elif user_table_choice == 12:
        game_score[12][player_col] = kind_4(dice_roll)
    else:
        game_score[13][player_col] = combo_yahtzee(dice_roll)
    
    return game_score
```
